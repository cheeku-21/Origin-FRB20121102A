```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("../datasets/properties.csv")
df.head()
```

```python
df = df.drop(columns=['Unnamed: 0'], axis=1)
df.head()
```

```python
import umap
import hdbscan
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.metrics import adjusted_rand_score

rand_scores = []
adjust_rand_scores = []
neighbor_pairs = []
n_neighbors_values = [5,6,7,8,9]
cluster_labels_by_n = {}
```

```python
clusterable_embedding_1 = umap.UMAP(
    n_neighbors=5,
    min_dist=0.01,
    n_components=2,
).fit_transform(df)

clusterer1 = hdbscan.HDBSCAN(min_cluster_size=80, min_samples=15)
cluster_labels1 = clusterer1.fit_predict(clusterable_embedding_1)

cluster_labels_by_n[5] = cluster_labels1

plt.figure(figsize=(8, 6))
sns.set(style="white", rc={"axes.facecolor": (0, 0, 0, 0)})

# Scatter plot with points colored by cluster labels
scatter = plt.scatter(clusterable_embedding_1[:, 0], clusterable_embedding_1[:, 1], c=cluster_labels1, cmap='Spectral', s=10)
plt.title(f"UMAP + HDBSCAN Clustering (n_neighbors = {5})")
plt.colorbar(scatter, label='Cluster Labels of 5 n_neighbour')
plt.show()
```

```python
clusterable_embedding_2 = umap.UMAP(
    n_neighbors=6,
    min_dist=0.01,
    n_components=2,
).fit_transform(df)

clusterer2 = hdbscan.HDBSCAN(min_cluster_size=80, min_samples=15)
cluster_labels2 = clusterer2.fit_predict(clusterable_embedding_2)

cluster_labels_by_n[6] = cluster_labels2

plt.figure(figsize=(8, 6))
sns.set(style="white", rc={"axes.facecolor": (0, 0, 0, 0)})

# Scatter plot with points colored by cluster labels
scatter = plt.scatter(clusterable_embedding_2[:, 0], clusterable_embedding_2[:, 1], c=cluster_labels2, cmap='Spectral', s=10)
plt.title(f"UMAP + HDBSCAN Clustering (n_neighbors = {6})")
plt.colorbar(scatter, label='Cluster Labels of 6 n_neighbour')
plt.legend(title="Clusters", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
clusterable_embedding_3 = umap.UMAP(
    n_neighbors=7,
    min_dist=0.01,
    n_components=2,
    # metric = "eucledian",
).fit_transform(df)

clusterer3 = hdbscan.HDBSCAN(min_cluster_size=80, min_samples=15)
cluster_labels3 = clusterer3.fit_predict(clusterable_embedding_3)

cluster_labels_by_n[7] = cluster_labels3

plt.figure(figsize=(8, 6))
sns.set(style="white", rc={"axes.facecolor": (0, 0, 0, 0)})

# Scatter plot with points colored by cluster labels
scatter = plt.scatter(clusterable_embedding_3[:, 0], clusterable_embedding_3[:, 1], c=cluster_labels3, cmap='Spectral', s=10)
plt.title(f"UMAP + HDBSCAN Clustering (n_neighbors = {7})")
plt.colorbar(scatter, label='Cluster Labels of 7 n_neighbour')
plt.show()
```

```python
clusterable_embedding_4 = umap.UMAP(
    n_neighbors=8,
    min_dist=0.01,
    n_components=2,
).fit_transform(df)

clusterer4 = hdbscan.HDBSCAN(min_cluster_size=80, min_samples=15)
cluster_labels4 = clusterer4.fit_predict(clusterable_embedding_4)

cluster_labels_by_n[8] = cluster_labels4

plt.figure(figsize=(8, 6))
sns.set(style="white", rc={"axes.facecolor": (0, 0, 0, 0)})

# Scatter plot with points colored by cluster labels
scatter = plt.scatter(clusterable_embedding_4[:, 0], clusterable_embedding_4[:, 1], c=cluster_labels4, cmap='Spectral', s=10)
plt.title(f"UMAP + HDBSCAN Clustering (n_neighbors = {8})")
plt.colorbar(scatter, label='Cluster Labels of 8 n_neighbour')
plt.show()
```

```python
clusterable_embedding_5 = umap.UMAP(
    n_neighbors=9,
    min_dist=0.01,
    n_components=2,
    # metric = "eucledian",
).fit_transform(df)

clusterer5 = hdbscan.HDBSCAN(min_cluster_size=80, min_samples=15)
cluster_labels5 = clusterer5.fit_predict(clusterable_embedding_5)

cluster_labels_by_n[9] = cluster_labels5

plt.figure(figsize=(8, 6))
sns.set(style="white", rc={"axes.facecolor": (0, 0, 0, 0)})

# Scatter plot with points colored by cluster labels
scatter = plt.scatter(clusterable_embedding_5[:, 0], clusterable_embedding_5[:, 1], c=cluster_labels5, cmap='Spectral', s=10)
plt.title(f"UMAP + HDBSCAN Clustering (n_neighbors = {9})")
plt.colorbar(scatter, label='Cluster Labels of 9 n_neighbour')
plt.show()
```

```python
for i in range(len(n_neighbors_values)):
    for j in range(i + 1, len(n_neighbors_values)):
        n1, n2 = n_neighbors_values[i], n_neighbors_values[j]
        labels1, labels2 = cluster_labels_by_n[n1], cluster_labels_by_n[n2]

        # Calculate scores
        rand_score = adjusted_rand_score(labels1, labels2)
        adjusted_rand_score_val = adjusted_rand_score(labels1, labels2)

        # Append results for plotting
        rand_scores.append(rand_score)
        adjust_rand_scores.append(adjusted_rand_score_val)
        neighbor_pairs.append(f"({n1}, {n2})")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(rand_scores, adjust_rand_scores, color='blue', s=50)

for i, pair in enumerate(neighbor_pairs):
    plt.annotate(pair, (rand_scores[i], adjust_rand_scores[i]), textcoords="offset points", xytext=(5,5), ha='center')

plt.xlabel("Rand Score")
plt.ylabel("Adjusted Rand Score")
plt.title("Rand Score vs Adjusted Rand Score for Different n_neighbors Pairs")
plt.show()
```
